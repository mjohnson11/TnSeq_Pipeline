```python
sd = pd.read_csv('sk_data_headers_changed.csv')
sd['new.num.sig'] = sd.apply(lambda row: len([i for i in [j for j in range(1, 21) if j not in [9, 18]] if row['sig.Seg' + str(i).zfill(2)]]), axis=1)
ann_dat = pd.read_csv('../BT_BC_Association/Annotation/BT_edges_annotated.csv')

assays = ['A1', 'A3']
libs = ['BT_N' + str(i) for i in range(1,11)] + ['BT_H' + str(i) for i in range(1,11)]
tps = [0, 1, 2, 3, 4]

plas_lib_info = pd.read_csv('scripts/Plasmid_Lib_Demult.csv')
pl3 = {i[1]: i[0] for i in plas_lib_info.as_matrix(['Lib 3', 'quick number'])}
pl1 = {i[1]: i[0] for i in plas_lib_info.as_matrix(['Lib 1', 'quick number'])}
pl_dict = {'A1': pl1, 'A3': pl3}

dats = dict()
for a in assays:
    for clone_num in range(1, 21):
        lib = pl_dict[a][clone_num]
        clone = str(clone_num)
        rep = str(clone) + '_' + a + '_' + lib[3:]
        dats[rep] = pd.read_csv('new_s_measurements/' + rep + '_bc_s.csv')
        
reps = sorted([i for i in dats.keys()])

all_edges = set()
initial_edge_s = dict()
for rep in reps:
    initial_edge_s[rep] = dict()
    all_edges.update(set(dats[rep]['Edge']))
    for edge in set(dats[rep]['Edge']):
        td = dats[rep].loc[dats[rep]['Edge']==edge]
        initial_edge_s[rep][edge] = [np.mean(td['mean.s']), np.median(td['mean.s']), np.std(td['mean.s']), len(td)]
```

```python
median_deviations = dict()
for rep in reps:
    median_deviations[rep] = dict()
    for edge in set(dats[rep]['Edge']):
        td = dats[rep].loc[dats[rep]['Edge']==edge]
        median_deviations[rep][edge] = [i - initial_edge_s[rep][edge][1] for i in list(td['mean.s'])]
        
excluded_clones = [9, 18]

excluded_bcs = dict()
with open('BT_outlier_BCs.csv', 'w') as outfile:
    writer = csv.writer(outfile)
    writer.writerow(['Rep', 'BC', 'Edge', 'mean.s', 'edge.median.s'])
    for rep in reps:
        excluded_bcs[rep] = []
        for edge in set(dats[rep]['Edge']):
            td = dats[rep].loc[dats[rep]['Edge']==edge]
            if len(td) > 2:
                median = initial_edge_s[rep][edge][1]
                ebc_rec = td.loc[np.abs(td['mean.s'] - median) > 0.035].as_matrix(['BC', 'Edge', 'mean.s'])
                for rec in ebc_rec:
                    excluded_bcs[rep].append(rec)
                    writer.writerow([rep] + [str(i) for i in rec] + [median])
                    
x_info = pd.read_csv('../../Background_Data_Explore/Kruglyak_experiment/LK_clones_chosen_8_28_17.csv')
segs_in_order = ['LK1-D06',
                'LK2-A12',
                'LK2-E07',
                'LK2-F11',
                'LK3-D08',
                'LK3-G02',
                'LK4-A04',
                'LK4-E02',
                'LK1-B05',
                'LK1-C09',
                'LK1-E09',
                'LK2-B04',
                'LK4-B12',
                'LK4-D01',
                'LK4-H11',
                'LK1-C11',
                'LK2-A01',
                'LK3-C01',
                'LK3-D09',
                'LK4-A12']
#print(x_info.keys())
seg_to_fit = {i[0]: i[1:] for i in x_info.as_matrix(['segregant', 'MJ.Mean.Fitness', 'initial fitness, YPD 30C'])}
clone_to_fit = {segs_in_order.index(i)+1: seg_to_fit[i][0] for i in seg_to_fit}
EJ_data_clone_to_fit = {segs_in_order.index(i)+1: seg_to_fit[i][1] for i in seg_to_fit}
#merging with annotation data
mg = sd.merge(ann_dat, on='Edge', how='left')
```

# OK on to simulating

```python
def wf_step(f, s, N):
    f_alt = f*np.exp(s)
    f_use = f_alt / sum(f_alt)
    new_counts = np.random.multinomial(N, f_use)
    return new_counts / sum(new_counts)
    

def simulate_evolution(initial_frequencies, s_vec, N_list, num_gens):
    f_rec = []
    current_f = initial_frequencies
    neut_s = np.array([0]*len(s_vec))
    for i in range(num_gens):
        if i == 0 or (i > 0 and (N_list[i] < N_list[i-1])):  #bottleneck step, no s effects here
            current_f = wf_step(current_f, neut_s, N_list[i])
        else:
            current_f = wf_step(current_f, s_vec, N_list[i])
        f_rec.append(current_f)
    return f_rec
```

* repeated 100,000 bottlenecks, with 10 gens growth

```python
%matplotlib inline
bottle = 100000
N_L = [1000*(2**i) for i in range(21)] + [bottle*(2**(i+3)) for i in range(8)] + [bottle*(2**i) for i in range(11)]*4
jnk = pl.semilogy(N_L)
```

```python
def run_sims(s_vals, reps):
    num_edges = len(s_vals)
    initial_freqs = np.array([1/num_edges]*num_edges)
    f_tests = [simulate_evolution(initial_freqs, s_vals, N_L, len(N_L)) for i in range(reps)]
    
    r_rec = []
    for f_test in f_tests:
        r_rec.append([np.random.multinomial(bottle, f_test[i]) for i in range(len(N_L)) if N_L[i] == (bottle*(2**10))])
    
    return r_rec
```

```python
sim_reads = dict()
s_vecs = dict()
edge_vecs = dict()
for i in range(1,21):
    if i not in excluded_clones:
        s_list = list(sig_choice['s.Seg' + str(i).zfill(2)].fillna(0)/100)
        s_list += list(neutral_choice['s.Seg' + str(i).zfill(2)].fillna(0)/100)
        s_list += [0, 0.04, 0.02]  # for HO, WHI2, SFL1
        edge_ids = list(sig_choice['ID'])
        edge_ids += list(neutral_choice['ID'])
        edge_ids += ['HO', 'SFL1', 'WHI2']
        # let's say each one gets barcoded 10 times
        s_list = s_list * 10
        edge_ids = edge_ids * 10
        s_vecs[i] = np.array(s_list)
        edge_vecs[i] = edge_ids
        sim_reads[i] = run_sims(np.array(s_list), 3)
```

```python
def calc_interval_logslope(row, t1, t2, sum1=bottle, sum2=bottle, cutoff=10):
    if row['T' + str(t1)] < cutoff or row['T' + str(t2)] < cutoff:
        return np.nan
    else:
        logdif = np.log(row['T' + str(t2)]/sum2) - np.log(row['T' + str(t1)]/sum1)
        return logdif / (10*(t2-t1))

def get_s_rec(row, cutoff=10):
    tmp_rec = []
    tps = [i for i in range(5) if row['T' + str(i)] >= cutoff]
    for i in range(len(tps)-1):
        t1 = tps[i]
        t2 = tps[i+1]
        tmp_rec.append(row['scaled.s_' + str(t1) + '_' + str(t2)])
    return tmp_rec

dfs = dict()
cols2 = ['mean.s', 'std.s', 'len.s']
neutral_indices = list(neutral_choice['ID']) + ['HO']
for rep in range(3):
    for i in range(1,21):
        if i not in excluded_clones:
            dfs[str(i)+'_'+str(rep)] = pd.DataFrame(np.array(sim_reads[i][0]).T, columns=['T0', 'T1', 'T2', 'T3', 'T4'])
            d = dfs[str(i)+'_'+str(rep)]
            d['True.s'] = s_vecs[i]
            d['Edge.index'] = edge_vecs[i]
            for i in range(4):
                for j in range(i+1, 5):
                    d['s_' + str(i) + '_' + str(j)] = d.apply(lambda r: calc_interval_logslope(r, i, j), axis=1)

            for i in range(4):
                for j in range(i+1, 5):
                    median_fit = np.nanmedian(d.loc[d['Edge.index'].isin(neutral_indices)]['s_' + str(i) + '_' + str(j)])
                    d['scaled.s_' + str(i) + '_' + str(j)] = d['s_' + str(i) + '_' + str(j)] - median_fit

            d['mean.s'] = d.apply(lambda r: np.nanmean(get_s_rec(r)), axis=1)
            d['std.s'] = d.apply(lambda r: np.nanstd(get_s_rec(r)), axis=1)
            d['len.s'] = d.apply(lambda r: len(get_s_rec(r)), axis=1)
```
